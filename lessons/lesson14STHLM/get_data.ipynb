```python
api_key = ""
movies_url = f"https://imdb-api.com/en/API/Top250Movies/{api_key}"
tv_url = f"https://imdb-api.com/en/API/Top250TVs/{api_key}"
import pandas as pd
import requests
```

```python
movie_res = requests.get(movies_url)

movie_data = movie_res.json()
```

```python
tv_res = requests.get(tv_url)

tv_data = tv_res.json()
```

Vi ska nu läsa in i dataframes

```python
movie_df = pd.DataFrame(movie_data["items"])
tv_df = pd.DataFrame(tv_data["items"])
movie_df
```

```python
movie_df.to_json("movies.json")
tv_df.to_json("tv.json")
```
