Läsa in våra json-filer

```python
import pandas as pd
movie_df = pd.read_json("movies.json")
tv_df = pd.read_json("tv.json")
movie_df
```

Best year for movies

```python
movie_df["year"].value_counts()
```

```python
top_year_movies = movie_df[movie_df["year"] == 1995]
top_year_movies
```

```python
tv_df["year"].value_counts()
```

```python
top_year_tv = tv_df[tv_df["year"] == 2015]
top_year_tv
```

Best decade for movies

```python
min = movie_df["year"].min()
max = movie_df["year"].max()
min, max
```

```python
start = round(min - 5, -1)
end = round(max + 5, -1)
start, end
```

```python
decades = {}
current = start

while current < end:
    df = movie_df[movie_df["year"].between(current, current + 9)]
    key = str(current) + "s"
    decades[key] = df
    current += 10

decades
```

```python
len(decades["2010s"])

for decade in decades.keys():
    print(decade, len(decades[decade]))
```

```python
decades["2000s"]
```

Vi hittar även bästa årtiondet för serier

```python
min = tv_df["year"].min()
max = tv_df["year"].max()
min, max
```

```python
start = round(min - 5, -1)
end = round(max + 5, -1)
start, end
```

```python
tv_decades = {}
current = start
while current < end:
    df = tv_df[tv_df["year"].between(current, current + 9)]
    key= str(current) + "s"
    tv_decades[key] = df
    current += 10
```

```python
tv_decades["1960s"]
```

```python
for tv_decade in tv_decades.keys():
    print(tv_decade, len(tv_decades[tv_decade]))
```

```python
tv_decades["2010s"]
```

Visualisera mera

```python
decades["2000s"].set_index("title").plot.pie(y="imDbRatingCount").legend('', frameon=False)
```
